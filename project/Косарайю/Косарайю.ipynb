```python
a_t = np.array(a).T
```

```python
visited=[False]*len(a)
order =[]
for vertex in range(len(a)):
    if not visited[vertex]:
        s1(vertex)
```

```python
component=[None]*len(a)
component_count=0
for vertex in reversed(order):
    if component[vertex] is None:
        s2(vertex)
        component_count+=1
```
